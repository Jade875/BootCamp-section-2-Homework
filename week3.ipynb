```python
#Question 3

def time_of_day_from_hour(hour):
  if 5 <= hour < 12:
    return 'morning'
  elif 12 <= hour < 17:
    return 'afternoon'
  elif 17 <= hour < 21:
    return 'evening'
  else:
    return 'night'

df['time_of_day']= df['hour'].apply(time_of_day_from_hour)
df[['hour', 'time_of_day']].head()

avg_peds_by_tod=df.groupby('time_of_day')['pedestrians'].mean()
print(avg_peds_by_tod)

plt.figure(figsize=(8,5))
sns.bar.plot(
    x=avg_peds_by_tod.index,
    y=avg_peds_by_tod.values
)
plt.title('Avg pedestrians count by time of day')
plt.xlabel('time of day')
plt.ylabel('avg pedestrian count')
plt.tight_layout()
plt.show()

```
